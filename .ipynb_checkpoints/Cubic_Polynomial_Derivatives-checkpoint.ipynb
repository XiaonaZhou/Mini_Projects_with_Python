# **Cubic Polynomial Derivatives**

#### A single-variable, cubic polynomial with integer coefficients is any function of the form $f(x)=ax^3+bx^2+cx+d$ where a, b, c, and d are integers with a $\neq$ 0. In this activity, we develop a function named cube_deriv, that plots the given cubic $f(x)$ in blue and its derivative $ f'(x)$ in red. Since x is between -10 and 10 and contains least 1000 points, I use list comprehension to assign values to x.  For the function cube_deriv, I use the basic rules for calculating derivative, which in this case is $3ax^2+2bx+c$. And, this formula is used in list comprehension to assign the y values. Similar process is used in assigning the values of $f(x)$. After all the values are obtained, I use plt.plot to plot the graphs, plt.label to add labels, and plt.show() to finally show the graphs.

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

i=[-10+ i*20/1000 for i in range(1001)]

def cube_deriv(a,b,c,d):
    return [3*a*x**(2)+2*b*x+c for x in i]
def ff(a,b,c,d):
    return [a*x**3+b*x**2+c*x+d for x in i]

plt.plot(i,cube_deriv(3,-2,0,5),label='cube_deriv')
plt.plot(i,ff(3,-2,0,5),label='f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs. y')
plt.legend()
plt.show()
```

```python
# Method 2 (use np.range instead of list comprehensions )
import matplotlib.pyplot as plt
import numpy as np
from math import *

x = np.arange(-10,10,0.02)

def cub_deriv(a,b,c,d):
    return 3*a*x**(2)+2*b*x+c
def ff(a,b,c,d):
    return a*x**3+b*x**2+c*x+d

plt.plot(x,cub_deriv(3,-2,0,5),label='cub_deriv')
plt.plot(x,ff(3,-2,0,5),label='f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs. y')
plt.legend()
plt.show()
```
