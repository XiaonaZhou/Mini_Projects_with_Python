# **An Adjustable Rate Mortgage**
An adjustable rate mortgage1(ARM) is a mortgage loan where the interest rate is adjusted from time to time based on some index. ARMs have changing interest rates that may make the loan more expensive or more favorable than a fixed rate mortgage. In this project we will compare a fixed rate mortgage to a â€œworst-case scenarioâ€ ARM. For a fixed rate mortgage, I use compound interst formula $ P'=P(1+r/n)+c $, where ð‘ƒ = Principal sum, ð‘Ÿ = annual interest rate, ð‘› = number of compoundings per year, ð‘ = contribution to principal per compounding
period, ð‘ƒ'= New principal sum after one compounding period.
For loan with fixed interest rate, I calculate the monthly payment with a while loop, and I use the monthly payment I get to calculate the amount the borrower pays to the bank with fixed rate using the function p_contributions_month(p, r, n, c, m).

For loan with ARM, I calculate the monthly payment during the first 10 years as if the loan were a 30-year fixed mortgage at 3.1%. I use the monthly payment I get to calculate the amount of principal left after 10 years. During the 11th year, the monthly payment is the same as if the loan were a 20-year fixed mortgage at 4.1%. I repeat a similar step to calculate the amount of principal left after the increment of interest rate. Lastly, I calculate the finally monthly payment for the remaining years as if the principal has to be zero by the end of 30 years period.

The amount the borrower pays to the bank if they accept the terms of Loan 1 is 378000 dollars, and Loan 2 is 417228 dollars. For loan 1, the monthly payment is 1050 dollars throught out the period of 30 years. In comparison, the monthly payment for loan 2 in the first ten years is only 855 dollars. However, as the interest rate increases, the monthly payment increases as well, Soon after 12th year, the monthly payment for loan 2 is greater than loan 1. As my caculation indicates, the total amount of money pay to the bank is different by 39228 dollars. One may consider take ARM is he or she is not consider to live in a house for more than 10 years or so, and he or she has the confidence to sell the property before that time. The fixed rate mortgage is a safer choice.
